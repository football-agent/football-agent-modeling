```python
#read the data
#goalkeepers = pd.read_csv("keepers-2.csv")
players1 = pd.read_csv("players-2.csv",low_memory=False)
#players2 = pd.read_csv("players_22.csv",low_memory=False)
players1 = pd.read_csv("top5_with_photos.csv",low_memory=False)
```

```python
#formatting
players1["age"] = players1["age"].str[:2]
players1["age"] =pd.to_numeric(players1["age"])
```

```python
def team_analysis(team): #analysis of team data over some seasons
    return players1[players1["squad"] == team].groupby(["season"]).describe().transpose()
    #TODO implment graphic

def player_analysis(player):
    return players1[players1["player"] == player].groupby("season").describe()
    #TODO extract only relevant information 

#def player_league(player):
#    for team in set(players1[players1["player"] == "Robert Lewandowski"]["squad"]):
        
```

```python
def team_stats(team,year,criteria):
    return players1[(players1["season"] >= year) &(players1["squad"] == team)][criteria].describe()

def team_stats_with_position(team,year,criteria):
    return players1[(players1["season"] >= year) &(players1["squad"] == team)][criteria].describe()

def team_stats2(team,year,criteria):
    median = players1[(players1["squad"]== team) & (players1["season"] >= year)]["games"].describe()["mean"]
    return players1[(players1["season"] >= year) &(players1["squad"] == team) &(players1["games"] > median)][criteria].describe()

def player_stats(player,year,criteria):
    return players1[(players1["player"] == player) & (players1["season"] >= year)][criteria].describe()

def comparison(player,team,year,criteria):
    print("Summary of the fit of",player,"to",team,"\n")
    print("Team-Wide Summary")
    for i in criteria:
        j=7
        while player_stats(player,year,criteria)[i]["max"] < team_stats(team,year,criteria)[i][j]:
            j = j-1
        if j ==7:
            print("You achieve the highest score regarding",i)
        elif j==6:
            print("You acieve a higher score than 75% regarding",i)
        elif (j==5) | (j==4):
            if player_stats(player,year,criteria)[i]["max"] in range(int(team_stats(team,year,criteria)[i][5]*1.1),int(team_stats(team,year,criteria)[i][5]*0.9)): #if its +/-10% from the average value
                print("You achieve an average score regarding",i)
            elif j==5:
                print("You achieve a higher score than average regarding",i)
            elif j==4:
                print("You achieve a lower score than average regarding",i)
        elif j==3:
            print("You achieve a lower score than 25% regarding",i)
        elif j==2:
            print("You achieve the lowest score regarding",i)
    print("\nComparison with Key players") # = those who are playing more than the median, TODO more than 10 games
    for i in criteria:
        j=7
        while player_stats(player,year,criteria)[i]["max"] < team_stats2(team,year,criteria)[i][j]:
            j = j-1
        if j ==7:
            print("You achieve the highest score regarding",i)
        elif j==6:
            print("You acieve a higher score than 75% regarding",i)
        elif (j==5) | (j==4):
            if player_stats(player,year,criteria)[i]["max"] in range(int(team_stats2(team,year,criteria)[i][5]*1.1),int(team_stats2(team,year,criteria)[i][5]*0.9)): #if its +/-10% from the average value
                print("You achieve an average score regarding",i)
            elif j==5:
                print("You achieve a higher score than average regarding",i)
            elif j==4:
                print("You achieve a lower score than average regarding",i)
        elif j==3:
            print("You achieve a lower score than 25% regarding",i)
        elif j==2:
            print("You achieve the lowest score regarding",i)

    print("\nPosition Specific Summary")
    for pos in ["DF","MF","FW"]:
        res = [i for i in set(players1[(players1["player"] == player) & (players1["season"] >= year)]["position"]) if pos in i] #will contain all strings with position value. Needed for case when player has multiple positions
        if (len(res) > 0) & (pos =="DF"):
            print("Defense Player")
            #lala
        if (len(res) > 0) & (pos =="MF"):
            print("Midfielder")
            #lala
        if (len(res) > 0) & (pos =="FW"):
            print("Forward")
            #lala
```
